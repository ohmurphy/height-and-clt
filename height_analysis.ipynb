# Heights and the Central Limit Theorem

### Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import norm
```

## Simple Random Sampling

### Sample from Normal

#### Individual Dist

```python
# Parameters for the normal distribution
mean = 170      # Mean (mu)
std_dev = 7     # Standard deviation (sigma)

# Create an array of x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF of the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Create the plot
plt.plot(x, pdf, color='g')
plt.title("Dist for an Individual's Height")
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.fill_between(x, pdf, alpha=0.5, color='g')
plt.savefig('individual_h_dist_normal.png', format='png')

# Show the plot
plt.show()
```

#### Population Dist

```python
np.random.seed(3)

# Parameters for the normal distribution
mean = 170      # Mean (mu)
std_dev = 7     # Standard deviation (sigma)
size = 100000    # Number of samples to draw

# Generate samples
samples = np.random.normal(loc=mean, scale=std_dev, size=size)

# Graph histogram
sns.histplot(samples, stat='density', color="r", alpha=0.7)
plt.title("Dist for Population Height")
plt.xlabel("Height (cm)")
plt.savefig('pop_h_dist_normal.png', format='png')
plt.show()
```

### Sample from Uniform

#### Individual Dist

```python
np.random.seed(1)

# Parameters for the uniform distribution
start = 140
end = 200

# Create an array of x values
x = np.linspace(start, end, 1000)

# Calculate the PDF of the normal distribution
pdf = np.ones(len(x)) / (end - start)

# Create the plot
plt.plot(x, pdf, color='g')
plt.title("Dist for an Individual's Height")
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.fill_between(x, pdf, alpha=0.5, color='g')
plt.savefig('individual_h_dist_uniform.png', bbox_inches='tight', format='png')

# Show the plot
plt.show()
```

#### Population Dist

```python
np.random.seed(4)

# Parameters for the uniform distribution
low = 140      # Lower bound
high = 200     # Upper bound
size = 100000  # Number of samples to draw

# Generate samples
samples = np.random.uniform(low=low, high=high, size=size)

# Graph histogram
sns.histplot(samples, stat='density', color="r", alpha=0.7)
plt.title("Dist for Population Height")
plt.xlabel("Height (cm)")
plt.savefig('pop_h_dist_uniform.png', bbox_inches='tight', format='png')
plt.show()
```

## Sampling Averages

### Individual Gene Dist

```python
# parameters for uniform dist, values set to get a dist centered at 170 w/ std deviation 7
low = 0              # lower bound on uniform dist
high = 12 * 49 / 340 # upper bound on uniform dist

x = np.linspace(low, high, 1000)

pdf = np.ones(len(x)) / (high - low)


# Graph histogram
plt.plot(x, pdf, color='b')
plt.title("Dist for Individual Gene Impact")
plt.xlabel("Height Impact (cm)")
plt.fill_between(x, pdf, alpha=0.5, color='b')
plt.savefig('individual_gene_averaging.png', format='png')
plt.show()
```

### Individual Distribution

```python
# Parameters for the normal distribution
mean = 170      # Mean (mu)
std_dev = 7     # Standard deviation (sigma)

# Create an array of x values
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF of the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Create the plot
plt.plot(x, pdf, color='g')
plt.title("Dist for an Individual's Height")
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.fill_between(x, pdf, alpha=0.5, color='g')
plt.savefig('individual_h_dist_normal.png', format='png')

# Show the plot
plt.show()
```

### Population Distribution

```python
np.random.seed(5)

# parameters for uniform dist, values set to get a dist centered at 170 w/ std deviation 7
low = 0              # lower bound on uniform dist
high = 12 * 49 / 340 # upper bound on uniform dist
n = 196              # num iid variables, e.g. genes, approximately 340 * 340 / (12 * 49)
num_samples = 100000
samples = np.empty(num_samples)
for i in range(num_samples):
    samples[i] = np.random.uniform(low=low, high=high, size=n).sum()

# Graph histogram
sns.histplot(samples, stat='density', color="r", alpha=0.7)
plt.title("Dist for Population Heights")
plt.xlabel("Height (cm)")
plt.savefig('pop_h_dist_averaging.png', format='png')
plt.show()
```

## Heritable Heights

```python
np.random.seed(9)

# parameters for uniform dist, values set to get a dist centered at 170 w/ std deviation 7
low = 0              # lower bound on uniform dist
high = 12 * 49 / 340 # upper bound on uniform dist
n = 196              # num iid variables, e.g. genes, approximately 340 * 340 / (12 * 49)
num_samples = 100000
samples = np.empty(num_samples)
mean_adjustment = 50
for i in range(num_samples):
    # evenly space the mean adjustments of the distributions
    samples[i] = np.random.uniform(low=low, high=high, size=n).sum() - mean_adjustment + 2 * mean_adjustment * (i / num_samples)

# Graph histogram
sns.histplot(samples, stat='density', color="r", alpha=0.7)
plt.title("Dist for Population Heights")
plt.xlabel("Height (cm)")
plt.savefig('pop_h_dist_heritable.png', format='png')
plt.show()
```

## Generational change

```python
np.random.seed(2)

# parameters for uniform dist, values set to get a dist centered at 170 w/ std deviation 7
low = 0              # lower bound on uniform dist
high = 12 * 49 / 340 # upper bound on uniform dist
n = 196              # num iid variables, e.g. genes, approximately 340 * 340 / (12 * 49)
num_samples = 100000
starting_pop_first_parent = np.empty(num_samples)
starting_pop_second_parent = np.empty(num_samples)
mean_adjustment = 50
for i in range(num_samples):
    starting_pop_first_parent[i] = np.random.uniform(low=low, high=high, size=n).sum() - mean_adjustment + 2 * mean_adjustment * (i / num_samples)
    starting_pop_second_parent[i] = np.random.uniform(low=low, high=high, size=n).sum() - mean_adjustment + 2 * mean_adjustment * (i / num_samples)

# Graph starting pop histogram
sns.histplot(np.append(starting_pop_first_parent, starting_pop_second_parent), color="y", stat='density', alpha=0.2, label="first")

hist_params = dict()
hist_params[2] = {"label": "second", "color": "g", "alpha": 0.5}
hist_params[10] = {"label": "tenth", "color": "b", "alpha": 0.8}

# shuffle pop to avoid sorting of parents
np.random.shuffle(starting_pop_first_parent)

# means for next generation are average of parent
gen_means = (starting_pop_first_parent + starting_pop_second_parent) / 2

for i in range(2, 101):
    gen_pop_first_partner = np.empty(num_samples)
    gen_pop_second_partner = np.empty(num_samples)

    std_dev = 7
    for j in range(num_samples):
        gen_pop_first_partner[j] = np.random.normal(loc=gen_means[j], scale=std_dev, size=1)
        gen_pop_second_partner[j] = np.random.normal(loc=gen_means[j], scale=std_dev, size=1)

    if i in hist_params.keys():
        params = hist_params[i]
        # Graph histogram for generation
        sns.histplot(np.append(gen_pop_first_partner, gen_pop_second_partner), stat='density', color=params["color"], alpha=params["alpha"], label=params["label"])
    
    # shuffle pop to avoid sorting of parents
    np.random.shuffle(gen_pop_first_partner)

    # means for next generation are average of parent
    gen_means = (gen_pop_first_partner + gen_pop_second_partner) / 2

plt.title("Dist for Population Heights")
plt.xlabel("Height (cm)")
plt.legend()
plt.savefig('pop_h_dist_multi_gen.png', format='png')
plt.show()

hist_params[100] = {"label": "hundredth", "color": "r", "alpha": 0.7}
params = hist_params[100]
sns.histplot(np.append(gen_pop_first_partner, gen_pop_second_partner), stat='density', color=params["color"], alpha=params["alpha"], label=params["label"])

plt.title("Dist for Population Heights")
plt.xlabel("Height (cm)")
plt.legend()
plt.savefig('pop_h_dist_100_gen.png', format='png')
plt.show()
```
